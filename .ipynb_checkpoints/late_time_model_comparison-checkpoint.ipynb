```python
import numpy as np
import matplotlib.pyplot as plt

from bayesn_model import SEDmodel
```

```python
model = SEDmodel(load_model='M20+TM22_85days_uRv')
```

```python
t = np.arange(-10, 85, 1)
ls = ['--', '-', '-.']
fig, axs = plt.subplots(2, 4, sharex=True, sharey=True, figsize=(12, 8))
bands = ['u_CSP', 'B_CSP', 'V_CSP', 'r_CSP2', 'i_CSP2', 'Y_WIRC', 'J_WIRC', 'H_WIRC']
for b_ind, b in enumerate(bands):
    ax = axs.flatten()[b_ind]
    for i, theta in enumerate([-1, 0, 1]):
        flux = model.simulate_light_curve(t, 1, [b], theta=theta, mu=0, eps=0, AV=0, del_M=0, mag=True, write_to_files=False)[0]
        ax.plot(t, flux[:, 0], color='b', ls=ls[i], label=rf'$\theta=${theta}')
        ax.set_title(rf'{b}')
ax.invert_yaxis()
handles, labels = ax.get_legend_handles_labels()
print(handles, labels)
axs[1, 3].legend(handles, labels)
fig.supxlabel('Phase')
fig.supylabel('Absolute magnitude')
```

```python
t = np.arange(-10, 100, 1)
fig, axs = plt.subplots(2, 4, sharex=True, sharey=True, figsize=(12, 8))
bands = ['u_CSP', 'B_CSP', 'V_CSP', 'r_CSP2', 'i_CSP2', 'Y_WIRC', 'J_WIRC', 'H_WIRC']
for b_ind, b in enumerate(bands):
    ax = axs.flatten()[b_ind]
    flux = model.simulate_light_curve(t, 100, [b], theta=0, mu=0, eps=None, AV=0, del_M=0, mag=True, write_to_files=False)[0]
    mean, std = flux.mean(axis=1), flux.std(axis=1)
    ax.plot(t, mean, color='b', ls='-', label=rf'Mean')
    ax.plot(t, mean - std, color='b', ls='--', label=rf'Intrinsic scatter')
    ax.plot(t, mean + std, color='b', ls='--')
    ax.set_title(rf'{b}')
ax.invert_yaxis()
handles, labels = ax.get_legend_handles_labels()
print(handles, labels)
axs[1, 3].legend(handles, labels)
fig.supxlabel('Phase')
fig.supylabel('Absolute magnitude')
```
